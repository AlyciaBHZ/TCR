```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

file_path = 'tst.csv'
data = pd.read_csv(file_path)

def count_non_empty(row):
    return row.count() - 1  # subtract 1 to exclude 'peptide' column which is always present

data['non_empty_count'] = data.apply(count_non_empty, axis=1)
#print the head of the data

train, test = train_test_split(data, test_size=0.2, stratify=data['non_empty_count'], random_state=42)

# To plot, comment out this line:
train = train.drop(columns=['non_empty_count'])
test = test.drop(columns=['non_empty_count'])

train.to_csv('tst.csv', index=False)
test.to_csv('val.csv', index=False)

data.shape[0],train.shape[0], test.shape[0]

(89,23)
```
